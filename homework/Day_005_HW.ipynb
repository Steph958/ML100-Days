```python
df05 = pd.DataFrame()
df07 = pd.DataFrame()
df10 = pd.DataFrame()
for i in app_train:
    df = pd.DataFrame()
    if (i != 'TARGET') and (app_train[i].dtype != 'object'):
        df['TARGET'] = app_train['TARGET']
        df[i] = app_train[i]
        coe = abs(df.corr().iloc[0,1])
        if coe > 0.05:
            df05[i] = df[i]
            if coe > 0.07:
                df07[i] = df[i]
                if coe > 0.1:
                    df10[i] = df[i]
        
# df = df.dropna(axis=0)
```

```python
print("df05:", df05.columns.values)
print()
print("df07:", df07.columns.values)
print()
print("df10:", df10.columns.values)
dffeature = list(df07)
dffeature
```

```python
feature = ['AMT_INCOME_TOTAL','AMT_CREDIT']+dffeature + ['TARGET']
df = app_train[feature]
```

```python
import seaborn as sns
f,ax = plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(),annot=True,linewidths=0.5, fmt='.2f', ax=ax)
plt.show()
```

```python
plt.hist(df['DAYS_BIRTH'],bins=15)
plt.xlim(df['DAYS_BIRTH'].min(),df['DAYS_BIRTH'].max())
plt.xlabel('DAYS_BIRTH')
```

```python
plt.hist(df['AMT_CREDIT'],bins=25)
plt.xlim(df['AMT_CREDIT'].min(),df['AMT_CREDIT'].max())
plt.xlabel('AMT_CREDIT')
```
